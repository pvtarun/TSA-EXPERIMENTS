```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dates = pd.date_range('2025-01-20', periods=5)  # Dates from 2025-01-20 to 2025-01-24
prices = [100]
```

```python
for _ in range(4):
    prices.append(prices[-1] + np.random.uniform(-2, 2))
```

```python
data = pd.DataFrame({'Date': dates, 'Stock_Price': prices})
```

```python
data.loc[data['Date'] == '2025-01-22', 'Date'] = np.nan
data.loc[data['Date'].isna(), 'Stock_Price'] = 102.25
```

```python
print("Data with Missing Date (NaN in the Date Column):")
print(data)
```

```python
data_interpolated = data.set_index('Date').interpolate(method='linear')
```

```python
print("\nData After Interpolation:")
print(data_interpolated)
```

```python
data_interpolated.plot(marker='o', title="Stock Price Over 5 Days (After Interpolation)", grid=True)
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.xticks(rotation=45)
plt.show()
```
