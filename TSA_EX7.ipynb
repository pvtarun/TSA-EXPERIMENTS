```python
# STEP 1: Install necessary packages (only for Colab)
!pip install statsmodels --quiet

# STEP 2: Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# STEP 3: Generate synthetic time series data
np.random.seed(42)
date_range = pd.date_range(start='2020-01-01', periods=36, freq='M')  # Monthly data for 3 years
trend = np.linspace(10, 50, 36)
seasonality = 10 * np.sin(np.linspace(0, 3 * 2 * np.pi, 36))  # seasonal effect
noise = np.random.normal(0, 2, 36)
data = trend + seasonality + noise

# STEP 4: Create time series DataFrame
ts_data = pd.DataFrame({'Date': date_range, 'Value': data})
ts_data.set_index('Date', inplace=True)

# STEP 5: Plot original time series
ts_data.plot(title='Original Time Series', figsize=(10, 4))
plt.grid()
plt.show()

# STEP 6: Decompose the time series
result = seasonal_decompose(ts_data['Value'], model='additive', period=12)

# STEP 7: Plot decomposed components
result.plot()
plt.suptitle('Time Series Decomposition', fontsize=16)
plt.tight_layout()
plt.show()
```
