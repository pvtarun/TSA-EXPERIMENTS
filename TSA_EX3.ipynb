```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a Large Time-Series Dataset
date_range = pd.date_range(start='2010-01-01', periods=1000, freq='D')  # 1000 days of data
np.random.seed(42)
values = np.random.randint(50, 200, size=len(date_range))  # Random values between 50 and 200

# Create a DataFrame
data = pd.DataFrame({'Date': date_range, 'Value': values})

# Save to CSV
data.to_csv('time_series_data.csv', index=False)
print("Dataset 'time_series_data.csv' created successfully!")

# Step 2: Load Time-Series Data
data = pd.read_csv('time_series_data.csv', parse_dates=['Date'], index_col='Date')

# Step 3: Compute Rolling Statistics
rolling_mean = data['Value'].rolling(window=30).mean()  # 30-day rolling mean
rolling_std = data['Value'].rolling(window=30).std()    # 30-day rolling standard deviation

# Step 4: Plot Original Data with Rolling Statistics
plt.figure(figsize=(12, 6))
plt.plot(data['Value'], label='Original Data', color='blue', alpha=0.6)
plt.plot(rolling_mean, label='Rolling Mean (30 days)', color='red', linestyle='dashed')
plt.plot(rolling_std, label='Rolling Std Dev (30 days)', color='green', linestyle='dotted')
plt.legend()
plt.title('Rolling Mean & Standard Deviation for Stationarity Check')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
```
