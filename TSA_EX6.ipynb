```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample time series dataset
dates = pd.date_range(start="2000-01-01", periods=120, freq='M')  # 10 years of monthly data
np.random.seed(42)
passenger_counts = np.random.randint(100, 500, size=len(dates)) + np.linspace(100, 500, len(dates))

df = pd.DataFrame({'Month': dates, 'Passengers': passenger_counts})
df.set_index('Month', inplace=True)

# Moving Average Smoothing Function
def moving_average(series, window):
    return series.rolling(window=window).mean()

# Apply Moving Average Smoothing
window_size = 12  # 12-month moving average
smoothed_series = moving_average(df['Passengers'], window_size)

# Plot the original data vs. smoothed data
plt.figure(figsize=(12,6))
plt.plot(df.index, df['Passengers'], label='Original Data', color='blue', alpha=0.5)
plt.plot(df.index, smoothed_series, label=f'{window_size}-Month Moving Average', color='red')
plt.xlabel('Year')
plt.ylabel('Passenger Count')
plt.title('Moving Average Smoothing on Sample Data')
plt.legend()
plt.show()
```
