```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dates = pd.date_range('2025-01-20', periods=5)
prices = [100]
```

```python
for _ in range(4):
    prices.append(prices[-1] + np.random.uniform(-2, 2))
```

```python
data = pd.DataFrame({'Stock_Price': prices}, index=dates)
```

```python
data.iloc[2] = np.nan
```

```python
print("Data with Missing Value:")
print(data)
```

```python
data_interpolated = data.interpolate(method='linear')
```

```python
print("\nData After Interpolation:")
print(data_interpolated)
```

```python
data_interpolated.plot(marker='o', title="Stock Price Over 5 Days (After Interpolation)", grid=True)
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.xticks(rotation=45)
plt.show()
```
